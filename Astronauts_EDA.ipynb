# 1. Data preparation / Cleaning

In this section, we are going to clean the two datasets for an optimal Exploratory Data Analysis

## 1.1. Data preparation for the table Astronauts

```python
# Importing the dataset
astro = pd.read_csv("astronauts.csv")

# Renaming "field21" with original name "eva"
astro = astro.rename(columns = {"field21": "eva"})

# Fixing astraunots with same id
astro['number'].loc[astro['name'] == 'Duque, Pedro'] = 386
astro = astro.copy()

# Droping useless columns
astro = astro.drop(['Unnamed: 0', 'id'], axis=1)
```

```python
astro.head()
```

###  Missing values:

```python
# Creating a small table summarising all missing values 

# Number of missing values for every columns
total = astro.isnull().sum().sort_values(ascending=False)

# Percentage of missing values for every columns 
percent = (astro.isnull().sum()/astro.isnull().count()).sort_values(ascending=False)

# Missing data
missing_data = pd.concat([total,percent],axis=1,keys=['Total','Pct'])

print( 'There is '+ str(len(astro)) + ' observations in this dataset.')
print('Here is the table of missions values:')
missing_data
```

There is a neglectable amount of missing values. We can drop NA safely without significantly affect our statistics:

```python
astro = astro.dropna()
```

In the dataset astronauts.csv, each line of represent a mission done by an astronaut. Astronauts can have done one or more missions. Thus, in order to conduct an exploratory data analysis (EDA) on the astronauts' table we should aggregate them.

```python
# Age
astro['age'] = astro['year_of_mission'] - astro['year_of_birth']

# Years of experience
astro['years_of_exp'] = astro['year_of_mission'] - astro['year_of_selection']
```

```python
to_group = astro[['number','name', 
                  'age', 'sex', 'year_of_mission', 'nationality',
                  'military_civilian', 'years_of_exp',
                  'mission_number', 'total_hrs_sum',
                  'total_eva_hrs','year_of_selection' ]]

# The table no_dup corresponds to the table Astronauts, without any duplicate (the key is the astronauts number)
no_dup = to_group.groupby('number').max()
no_dup = no_dup.reset_index()
no_dup = no_dup.rename(columns={'mission_number': 'nof_missions'})
```

## 1.2. Data preparation for the table Missions

```python
# Importing csv
missions = pd.read_csv("missions.csv")
missions = missions.drop(['Unnamed: 0', 'ID'], axis=1)

# Renaming " Cost" with "Cost"
missions = missions.rename(columns = {" Cost": "Cost"})

# Turning cost from object to float
missions['Cost'] = missions['Cost'].str.replace(',', "")
missions['Cost'] = missions['Cost'].apply(float)

# Creating feature year_of_mission
missions['year_of_mission'] = missions['Date'].apply(lambda x: re.findall('(\d{4})', x)[0])
```

### Missing values: 

```python
# Creating a small table sumarizing all missing values 

# Number of missing values for every columns
total = missions.isnull().sum().sort_values(ascending=False) 

# Percentage of missing values for every columns 
percent=(missions.isnull().sum()/missions.isnull().count()).sort_values(ascending=False) 

# Missing data
missing_data=pd.concat([total,percent],axis=1,keys=['Total','Pct'])

print( 'There is '+ str(len(missions)) + ' observations in this dataset.')
print('Here is the table of missions values:')
missing_data
```

More than 70% of Cost are not detailed here. We cannot delete the missing observations as it represents a big proportion of the dataset. As a consequence, we chose not to retrieve any insight from this column in our data analysis. A next step would be to try to predict the cost, using Machine Learning algorithms.

## 1.3. Joining both tables for maximum information

To optimise our Exploratory Data Analysis, we concatenated both tables in order to feed the table Astronauts with additional information given in the table Missions. However, there was no explicit join key between both tables. We managed to create one key from the column "Detail" in the table Missions that we joined with "ascend_shuttle" (in the table Astronauts), since the spacecraft name is unique for each mission.

List of modifications to create key:
* In both table there are soft hyphens we had to remove
* Sometimes there are spaces or "-" that are not always at the same place in both key
* In the column Detail there is a letter "R"  that has nothing to do here for some spacecraft

```python
# Deleting spaces, "-" and soft hyphen
astro['ascend_shuttle']  = astro['ascend_shuttle'].apply(lambda x: x.replace(" ","").replace("-","").replace('\xad', ''))

#Splitting Details in two parts
missions['details_split_1'] = missions['Detail'].str.split("|",expand = True)[0]
missions['details_split_2'] = missions['Detail'].str.split("|", expand = True)[1]
missions['details_split_2'] = missions['details_split_2'].apply(lambda x: x[1:] if x.startswith(' ') else x)

# Deleting space and '-' and soft hyphen
missions['details_split_2']  = missions['details_split_2'].apply(lambda x: x.replace(" ","").replace("-","").replace('\xad', ''))

# Deleting parentesis and content in it
missions['details_split_2'] = missions['details_split_2'].str.replace(r"\(.*?\)","")

# Deleting letter R in the end of key
missions['details_split_2'] = missions['details_split_2'].apply(lambda x: x[:-1] if x[-1] == "R" else x)
```

```python
# Merging astraunauts and missions
astro_missions = astro.merge(missions, left_on='ascend_shuttle', right_on='details_split_2', how="left")
```

```python
# Number of unmached data
len(astro_missions[astro_missions['details_split_2'].isnull()])
```

```python
len(missions[missions['details_split_2'].str.startswith('SoyuzT')])
```

```python
len(astro[astro['ascend_shuttle'].str.startswith('SoyuzT')])
```

Most of the unmatched observations come from the fact that there are a lot of Soyuz type spacecraft that are missing in the missions table. This is a non neglectable of amount that we are loosing but we still have enough to provide robust estimation since we are not trying to build a machine-learning model but we are doing an EDA.
